```python
def get_veg_data():
    import requests
    from bs4 import BeautifulSoup
    a = "https://vegetablemarketprice.com"
    url = "https://vegetablemarketprice.com/market/andhraPradesh/today"
    req = requests.get("https://vegetablemarketprice.com/market/andhraPradesh/today")
    soup=BeautifulSoup(req.content,"html.parser")
    img = []
    for i in soup.find_all("div",{"class":'vegetableImageDiv'}):
        img.append(a+i.get("image-src-file"))

    imgs = []
    for i in range(len(img)):
        imgs.append(img[i].split("-")[0]+"-"+img[i].split("-")[-1].replace("64","256"))



    import pandas as pd
    df = pd.read_html("https://vegetablemarketprice.com/market/andhraPradesh/today")
    df = df[0]
    df.drop("Unnamed: 0",axis = 1,inplace = True)
    D = []
    for i in range(len(df)):
        D.append((df.iloc[i].to_dict()))
        d = {}
    d['vegerables'] = D

    for i,j in zip(d["vegerables"],imgs):
        i['Image'] = j
    return d
```
